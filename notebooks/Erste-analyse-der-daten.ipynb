```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------
# lengths = [...]  # Hier Ihre echten Daten einfügen
# ---------------------------------------------------------

# Statistiken berechnen
q1, median, q3 = np.percentile(lengths, [25, 50, 75])
min_val, max_val = min(lengths), max(lengths)
mean_val = np.mean(lengths)

# Logarithmische Einteilung der Bins:
# Wir gehen hier von "mindestens 1" bis "max_val".
# Falls min_val < 1, kann man den Start per log10(1) definieren,
# damit es keine Fehler gibt.
n_bins = 30  # Anzahl der log-Bins
bins = np.logspace(np.log10(max(1, min_val)), np.log10(max_val+1), n_bins)

fig, ax = plt.subplots(figsize=(10, 6))

# Histogramm mit log-Bins
counts, bin_edges, patches = ax.hist(
    lengths,
    bins=bins,
    edgecolor="black",
    alpha=0.7
)

ax.set_title("Histogramm der Textlängen (inkl. Außenseiter)", fontsize=14)
ax.set_xlabel("Textlänge (Anzahl Zeichen) – log-Skala")
ax.set_ylabel("Anzahl der Texte")

# Beschriftung der Balken in kleinerer Schrift
for count, left_edge, right_edge in zip(counts, bin_edges[:-1], bin_edges[1:]):
    if count > 0:
        bar_center = (left_edge + right_edge) / 2
        # An dieser Stelle kann es bei sehr engen Bins zu Überlappungen kommen.
        # Schriftgröße bewusst kleiner gewählt:
        ax.text(bar_center, count, int(count),
                ha='center', va='bottom', fontsize=8, rotation=90)

# X-Achse auf logarithmisch setzen
ax.set_xscale('log')

# Optional: leichten Winkel für die Tick-Labels
plt.xticks(rotation=30)

# Vertikale Linien für Min, Median, Mean und Max
# - Um verschiedene Farben zu nutzen, z. B. blau (Min), rot (Median), grün (Mean), magenta (Max)
ax.axvline(min_val, color='blue', linestyle='--', label=f"Min: {min_val}")
ax.axvline(median, color='red', linestyle='--', label=f"Median: {median:,.0f}")
ax.axvline(mean_val, color='green', linestyle='--', label=f"Mean: {int(mean_val):,}")
ax.axvline(max_val, color='magenta', linestyle='--', label=f"Max: {max_val:,}")

ax.legend(loc='best')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn-Stil aktivieren
sns.set_style("whitegrid")  


# --- Statistiken ---
q1, median, q3 = np.percentile(lengths, [25, 50, 75])
min_val, max_val = min(lengths), max(lengths)
mean_val = np.mean(lengths)

# --- Log-Binning einrichten ---
n_bins = 30  # Anzahl der Bins in log-Space
bins = np.logspace(np.log10(max(1, min_val)), np.log10(max_val + 1), n_bins)

# --- Plot anlegen ---
fig, ax = plt.subplots(figsize=(10, 6))

# Histogramm mit log-Bins
counts, bin_edges, patches = ax.hist(
    lengths,
    bins=bins,
    edgecolor="black",
    alpha=0.7,
    color="#69b3a2"   # leicht anderer Blauton
)

ax.set_title("Histogramm der Textlängen (inkl. Außenseiter)", fontsize=14)
ax.set_xlabel("Textlänge (Zeichen) – log-Skala")
ax.set_ylabel("Anzahl der Texte")

# Achsenskalierung: X = log
ax.set_xscale('log')

# Ggf. hilft ein leichtes Rotieren, damit die Labels nicht überlappen
plt.xticks(rotation=30)

# --- Balkenhöhen (counts) beschriften ---
# Wir verwenden hier z.B. 8pt, damit es nicht zu eng wird.
for count, left_edge, right_edge in zip(counts, bin_edges[:-1], bin_edges[1:]):
    if count > 0:
        bar_center = (left_edge + right_edge) / 2
        # Tausender-Trennzeichen bei der Anzahl:
        text_label = f"{int(count):,}".replace(",", ".")  
        # Rotation=90 vermeidet Überlappung bei schmalen Bins
        ax.text(bar_center, count, text_label,
                ha='center', va='bottom',
                fontsize=8, rotation=90,
                color="black")

# --- Vertikale Linien für Kenngrößen ---
# Zusätzliche Linien für Q1 und Q3
ax.axvline(min_val, color='blue', linestyle='--', label=f"Min: {min_val:,}")
ax.axvline(q1, color='orange', linestyle='--', label=f"Q1: {int(q1):,}")
ax.axvline(median, color='red', linestyle='--', label=f"Median: {int(median):,}")
ax.axvline(mean_val, color='green', linestyle='--', label=f"Mean: {int(mean_val):,}")
ax.axvline(q3, color='orange', linestyle=':', label=f"Q3: {int(q3):,}")
ax.axvline(max_val, color='magenta', linestyle='--', label=f"Max: {max_val:,}")

# Legenden-Optimierung
ax.legend(loc='upper left', fontsize=9)

# Engeres Layout, damit nichts abgeschnitten wird
plt.tight_layout()
plt.show()
```
